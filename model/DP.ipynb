```python
import pandas as pd
from pandas import DataFrame,Series
import matplotlib.pyplot as plt
```

```python
col = ['frame','pre_num','pre_oflow','cur_num','cur_inflow']
data_1949 = pd.read_csv('../data/sample/1949_node_sample.txt',delim_whitespace=True,names=col)
data_1951 = pd.read_csv('../data/sample/1951_node_sample.txt',delim_whitespace=True,names=col)
data_2073 = pd.read_csv('../data/sample/2073_node_sample.txt',delim_whitespace=True,names=col)
data_2075 = pd.read_csv('../data/sample/2075_node_sample.txt',delim_whitespace=True,names=col)
#data[data.pre_oflow != data.cur_inflow]
```

```python
fig = plt.figure()
fig.subplots(2,2)
ax1 = fig.add_subplot(2,2,1) 
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
data_1949.pre_oflow.plot( ax = ax1)
data_1951.cur_inflow.plot( ax = ax2)
data_2073.cur_inflow.plot( ax = ax3)
data_2075.cur_inflow.plot( ax = ax4)
plt.show()
```

```python
data_1951[['pre_num','cur_num','pre_oflow']].plot()
plt.show()
```
